## Standard FLRW Friedmann Equation

$$
H^2 = \left( \frac{\dot{a}}{a} \right)^2 = \frac{8\pi G}{3} \rho
$$

**Units:**
- $H$: $[T^{-1}]$ (inverse time)
- $H^2$: $[T^{-2}]$
- $\rho$: $[M L^{-3}]$ (mass density)
- $G$: $[M^{-1} L^3 T^{-2}]$ (gravitational constant)

**Check:** $\frac{G \rho}{1} = [M^{-1} L^3 T^{-2}] \times [M L^{-3}] = [T^{-2}]$ ✓

## Spherical Multipole Expansion (Analogy)

In electrostatics, potential at distance $r$ from a source:

$$
\Phi(r, \theta, \phi) = \frac{1}{4\pi\epsilon_0} \left[ \frac{Q}{r} + \frac{\vec{p} \cdot \hat{r}}{r^2} + \frac{Q_{ij} r_i r_j}{r^3} + \ldots \right]
$$

**Units:**
- Monopole: $Q / r$ → $[Q]$ must have units $[\text{charge} \times \text{length}] / [\text{length}] = [\text{charge}]$
- Dipole: $\vec{p} / r^2$ → $[\vec{p}]$ must have units $[\text{charge} \times \text{length}^2] / [\text{length}^2] = [\text{charge} \times \text{length}]$
- Quadrupole: $Q_{ij} / r^3$ → $[Q_{ij}]$ must have units $[\text{charge} \times \text{length}^3] / [\text{length}^3] = [\text{charge} \times \text{length}^2]$

**Pattern:** Each higher multipole introduces one additional power of length.

## Gravitational Analogue

For gravity, replace charge with mass:

$$
\Phi_G(r) = -\frac{G M}{r} - \frac{G D}{r^2} - \frac{G Q}{r^3} + \ldots
$$

**Units:**
- $\Phi_G$: $[L^2 T^{-2}]$ (gravitational potential)
- $G M / r$: $[M^{-1} L^3 T^{-2}] \times [M] / [L] = [L^2 T^{-2}]$ ✓
- $G D / r^2$: Must have same units → $[D] = [M \times L]$
- $G Q / r^3$: Must have same units → $[Q] = [M \times L^2]$

**Required units for multipole moments:**

| Multipole | Symbol | Required Units |
|-----------|--------|----------------|
| Monopole | $M$ | $[M]$ (mass) |
| Dipole | $D$ | $[M \times L]$ (mass × length) |
| Quadrupole | $Q$ | $[M \times L^2]$ (mass × length²) |

## Symbolic Dimensional Check (Python)

```python
import sympy as sp
from sympy.physics.units import length, mass, time, gravitational_constant

# Define symbolic quantities
M = sp.Symbol('M', positive=True)  # mass
L = sp.Symbol('L', positive=True)  # length
T = sp.Symbol('T', positive=True)  # time

# Hubble parameter
H = 1 / T  # [T^-1]
H_squared = H**2  # [T^-2]

print("Hubble parameter H: [T^-1]")
print("H^2: [T^-2]")
print()

# Gravitational constant
G_units = L**3 / (M * T**2)  # [M^-1 L^3 T^-2]
print(f"Gravitational constant G: [M^-1 L^3 T^-2]")
print()

# Energy density
rho_units = M / L**3  # [M L^-3]
print(f"Energy density rho: [M L^-3]")
print()

# Check Friedmann equation dimensions
friedmann_lhs = H_squared
friedmann_rhs = G_units * rho_units

print("Friedmann equation dimensional check:")
print(f"  LHS (H^2): {friedmann_lhs}")
print(f"  RHS (G*rho): {sp.simplify(friedmann_rhs)}")
print(f"  Match: {sp.simplify(friedmann_lhs - friedmann_rhs) == 0}")
print()
```

```python
# Multipole moment units
print("Required units for multipole moments in gravitational potential:")
print()

# Monopole: Phi ~ G*M/r
monopole_numerator = M  # [M]
print(f"  Monopole moment: [M]")
print()

# Dipole: Phi ~ G*D/r^2
# Required: G*D/r^2 has same units as G*M/r
# [M^-1 L^3 T^-2] * [D] / [L^2] = [M^-1 L^3 T^-2] * [M] / [L]
# [D] / [L^2] = [M] / [L]
# [D] = [M * L]
dipole_numerator = M * L  # [M * L]
print(f"  Dipole moment: [M * L]")
print()

# Quadrupole: Phi ~ G*Q/r^3
# [Q] / [L^3] = [M] / [L]
# [Q] = [M * L^2]
quadrupole_numerator = M * L**2  # [M * L^2]
print(f"  Quadrupole moment: [M * L^2]")
print()
```

```python
# MULTING beta parameters: dimensional analysis
print("MULTING beta parameters — two interpretations:")
print()

# Interpretation 1: beta_d has units [L^2]
print("Interpretation 1: Dimensional betas")
beta_d_dim = L**2  # [L^2]
beta_q_dim = L**4  # [L^4]
print(f"  beta_d: [L^2] (e.g., 4.25 Mpc^2)")
print(f"  beta_q: [L^4] (e.g., 8.10 Mpc^4)")
print()

# Dipole term: beta_d / r(z)^2
r_z = L  # comoving distance
f_d_dim = beta_d_dim / r_z**2  # [L^2] / [L^2] = dimensionless
print(f"  Dipole term f_d = beta_d / r(z)^2: {sp.simplify(f_d_dim)} (dimensionless) ✓")
print()

# Quadrupole term: beta_q / r(z)^4
f_q_dim = beta_q_dim / r_z**4  # [L^4] / [L^4] = dimensionless
print(f"  Quadrupole term f_q = beta_q / r(z)^4: {sp.simplify(f_q_dim)} (dimensionless) ✓")
print()

print("Interpretation 2: Dimensionless betas with hidden reference scale")
beta_d_dimless = sp.Symbol('beta_d', real=True)  # dimensionless
beta_q_dimless = sp.Symbol('beta_q', real=True)  # dimensionless
r_A = sp.Symbol('r_A', positive=True)  # reference length [L]
r_P = sp.Symbol('r_P', positive=True)  # reference length [L]

print(f"  beta_d: dimensionless (e.g., 0.78)")
print(f"  beta_q: dimensionless (e.g., 0.19)")
print(f"  r_A: hidden reference length [L] for dipole")
print(f"  r_P: hidden reference length [L] for quadrupole")
print()

# Dipole term: beta_d * (r_A / r(z))^2
f_d_dimless = beta_d_dimless * (r_A / r_z)**2  # dimensionless
print(f"  Dipole term f_d = beta_d × (r_A / r(z))^2: dimensionless ✓")
print()

# Quadrupole term: beta_q * (r_P / r(z))^4
f_q_dimless = beta_q_dimless * (r_P / r_z)**4  # dimensionless
print(f"  Quadrupole term f_q = beta_q × (r_P / r(z))^4: dimensionless ✓")
print()
```
